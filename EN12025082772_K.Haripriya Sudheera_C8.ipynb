```python
import scipy.stats as stats
import numpy as np
# Given data
sample_mean = 2.8
population_mean = 3.0
sample_std_dev = 0.5
n = 50
alpha = 0.05
# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(n))
# Degrees of freedom
df = n - 1
# Critical t-value for one-tailed test
critical_value = stats.t.ppf(1 - alpha, df)
# Print results
print(f"T-statistic: {t_statistic}")
print(f"Critical value: {critical_value}")
# Conclusion
if t_statistic < critical_value:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"
print("Conclusion:", conclusion)
```

```python
#Task2:
from math import factorial
# Function to calculate combinations
def combinations(n, k):
    if k > n or k < 0:
        return 0
    return factorial(n) // (factorial(k) * factorial(n - k))
# Given data
total_families = 95
small_families = 55
large_families = total_families - small_families
families_to_select = 20

# Calculate different cases using the custom combinations function
case_1 = combinations(small_families, 18) * combinations(large_families, 2)   # Case: 18 small, 2 large
case_2 = combinations(small_families, 19) * combinations(large_families, 1)   # Case: 19 small, 1 large
case_3 = combinations(small_families, 20)                              # Case: 20 small

# Total ways to select families
total_ways = case_1 + case_2 + case_3
# Print results
print("Total ways to select families:", total_ways)
```

```python
#Section_4 Task1:
import numpy as np
from scipy import stats

# Sample data
sample_before = [75, 80, 70, 85, 65, 90, 80, 75, 85, 80]
sample_after = [80, 85, 75, 90, 70, 95, 85, 80, 90, 85]

# Calculate the differences
differences = np.array(sample_after) - np.array(sample_before)

# Perform the paired t-test
t_statistic, p_value = stats.ttest_rel(sample_after, sample_before)

# Significance level
alpha = 0.01

# Print results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Conclusion
if p_value < alpha:
    conclusion = "Reject the null hypothesis"
else:
    conclusion = "Fail to reject the null hypothesis"

print("Conclusion:", conclusion)
```

```python
# Display the first few rows of the dataset to understand its structure
print(Gamessales_data.head())

# Step 2: Extract 'Global_Sales_M$' column
global_sales = Gamessales_data['Global_Sales_M$']

# Parameters for sampling
sample_size = 1000  # Number of samples to draw for each sample
num_samples = 30    # Number of samples to take

# Random sampling with replacement
samples = [np.random.choice(global_sales, size=sample_size, replace=True) for _ in range(num_samples)]

# Calculate sample means
sample_means = [np.mean(sample) for sample in samples]

# Step 3: Plotting the distribution of sample means
plt.figure(figsize=(12, 6))
sns.histplot(sample_means, bins=30, kde=True)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Means')
plt.ylabel('Frequency')
plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=1, label='Mean of Sample Means')
plt.axvline(np.mean(global_sales), color='green', linestyle='dashed', linewidth=1, label='Population Mean')
plt.legend()
plt.show()

# Calculate statistics for sample means
mean_of_sample_means = np.mean(sample_means)
std_of_sample_means = np.std(sample_means)

print(f'Mean of Sample Means: {mean_of_sample_means}')
print(f'Standard Deviation of Sample Means: {std_of_sample_means}')
```
